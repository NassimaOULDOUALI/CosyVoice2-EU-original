### Run to create full demo data for webpage:

```python

SETTINGS = [
    "llm"
    , "flow"
    , "hifigan"
    , "llm_flow"
    , "llm_hifigan"
    , "flow_hifigan"
    , "llm_flow_hifigan"
    , "original"
]
```

```python
# fine-tune models:
model_dir = "pretrained_models/CosyVoice2-0.5B"
backbone = "blanken" # one of: "blanken", "hf:Qwen/Qwen2.5-0.5B", "hf:Qwen/Qwen3-0.6B", "hf:utter-project/EuroLLM-1.7B-Instruct"

llm_run_id = "3000-averaged-bl-mix"
flow_run_id = "3000-averaged-bl-mix"
hifigan_run_id = "original-real"

# settings
text_frontend = False
final = None
stream = False
speed = 1.0


for LANGUAGE in ["fr", "de"]:
    if LANGUAGE == "fr":
        text = "Parle en français. <|endofprompt|> Bonjour, je m'appelle Luka et je travaille dans une entreprise de technologie à Paris. Aujourd'hui, nous allons explorer les capacités de synthèse vocale en français avec CosyVoice 2."

    if LANGUAGE == "de":
        text = "<|de|><|endofprompt|> Guten Tag, mein Name ist Hans und ich arbeite in einem Technologieunternehmen in Berlin. Heute werden wir die Sprachsynthesefähigkeiten von CosyVoice 2 im Deutschen erkunden."

    # prompt_wav = "prompt_audio/macron.wav" if LANGUAGE == "fr" else "prompt_audio/merz.wav"

    prompt_wav = "prompt_audio/common_voice_fr_40952142.wav" if LANGUAGE == "fr" else "prompt_audio/common_voice_de_41123857.wav"


    for setting in SETTINGS:

        output_wav = f"audio_output/demo/{setting}-{LANGUAGE}.wav".replace("llm","slm")

        out_wav = run_inference(
            text=text,
            prompt_wav=prompt_wav,
            output_wav=output_wav,
            model_dir=model_dir,
            setting=setting,
            llm_run_id=llm_run_id,
            flow_run_id=flow_run_id,
            hifigan_run_id=hifigan_run_id,
            final=final,
            stream=stream,
            speed=speed,
            text_frontend=text_frontend,
            backbone=backbone,
        )
```
